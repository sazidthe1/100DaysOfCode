### **Big O notation üìà**

Big O notation is a mathematical notation used in computer science to describe the performance or complexity of an algorithm. It represents the upper bound or worst-case scenario of the time or space complexity of an algorithm as the input size increases.

### **Purpose of Big O notation ‚ú®**

The purpose of Big O notation is to provide a standardized way of analyzing and comparing the efficiency of algorithms. It allows us to understand how an algorithm's performance scales with larger input sizes and helps us make informed decisions when selecting algorithms for solving specific problems.

### **Advantages of Big O notation ‚úÖ**

üîπ **Standardized Comparison**: Big O notation provides a common language for expressing the efficiency of algorithms, allowing for easy comparison and evaluation of different approaches.

üîπ **Algorithm Selection**: Big O notation helps in selecting the most efficient algorithm for a specific problem by considering its time and space complexity.

üîπ **Efficiency Analysis**: Big O notation helps in analyzing the efficiency of an algorithm by providing insights into its growth rate as the input size increases.

üîπ **Scalability Prediction**: Big O notation allows us to predict the performance of an algorithm for larger input sizes without actually running the code.

üîπ **Communication and Collaboration**: Big O notation serves as a universal language for discussing and communicating algorithmic complexity among developers, researchers, and professionals in the field of computer science.

By leveraging Big O notation, we can analyze, compare, and select algorithms based on their efficiency. It helps in understanding the scalability and performance limitations of algorithms, aiding in the design and optimization of software systems.

#### **Traditional Approach**

```python
import time

# Start the timer
start_time = time.time()

empty_list = []

for i in range(10000000):
  empty_list.append(i)

# End the timer
end_time = time.time()

# Calculate the execution time
execution_time = end_time - start_time

# Print the execution time
print("Execution time:", execution_time, "seconds")
```

Rules:

##### **Here, time = a * n¬≤ + b ‚Üí O(n¬≤)**

```python
numbers = [3,5,2,7,4,3,6,7,8,9]

for i in range(len(numbers)):
  for j in range(i+1, len(numbers)):
    if numbers[i] == numbers[j]:
      print(str(numbers[i]) + " is a duplicate value!")
      break
```

##### **Here, time = a * n¬≤ + b * n + c ‚Üí O(n¬≤)**

```python
numbers = [3,5,2,7,4,3,6,7,8,9]
duplicates = []

for i in range(len(numbers)):             # n¬≤ iterations
  for j in range(i+1, len(numbers)):
    if numbers[i] == numbers[j]:
      duplicates = numbers[i]
      break

for i in range(len(numbers)):             # n iterations
  if numbers[i] == duplicates:
    print(str(i) + " is a duplicate value!")
```
