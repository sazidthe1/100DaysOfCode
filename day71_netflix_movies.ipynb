### Step 1: Loading and inspecting the dataset

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the CSV file
netflix_df = pd.read_csv('netflix_data.csv')
```

```python
# Checking the first 5 rows in the dataset
netflix_df.head()
```

### Step 2: Subsetting the DataFrame (Type - Movies Only)

```python
# Creating a subset for the netflix_df where the type is a "Movie"
netflix_subset = netflix_df[netflix_df["type"] == "Movie"]
```

### Step 3: Subsetting the columns of the new DataFrame (netflix_movies)

```python
# Subsetting the DataFrame again with specific columns
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]]
```

```python
# Inspecting the new dataframe
netflix_movies
```

### Step 4: Filtering the DataFrame by Movie duration

```python
# Filtering the DataFrame where movie duration < 60 minutes
short_movies = netflix_movies[netflix_movies['duration'] < 60]
```

```python
# Print the first 20 rows of DataFrame
short_movies.head(20)
```

### Step 5: Assigning colors to Movie genres

```python
# Creating a list to store colors for different genres
colors = []
```

```python
# Iterating through netflix_movies
for lab, row in netflix_movies.iterrows():
    if row['genre'] == "Children":
        colors.append("Blue")
    elif row['genre'] == "Documentaries":
        colors.append("Green")
    elif row['genre'] == "Stand-Up":
        colors.append("Orange")
    else:
        colors.append("Red")
```

### Step 6: Plotting the Movie duration by Year (using the genre colors created)

```python
# Initializing a figure object with the specified dimensions
fig = plt.figure(figsize=(12, 8))
```

```python
# Creating a list to store colors for different genres
colors = []

# Iterating through netflix_movies
for lab, row in netflix_movies.iterrows():
    if row['genre'] == "Children":
        colors.append("Blue")
    elif row['genre'] == "Documentaries":
        colors.append("Green")
    elif row['genre'] == "Stand-Up":
        colors.append("Red")
    else:
        colors.append("Gray")

# Initialize a figure object with the specified dimensions
fig = plt.figure(figsize=(12, 8))

# Create a scatter plot for movie duration by release year with colors
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)

# Add labels and title
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie Duration by Year of Release")

# Show the plot
plt.show()
```
