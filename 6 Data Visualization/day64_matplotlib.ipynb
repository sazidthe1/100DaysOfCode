### Matplotlib - Scatter, Bars, Histograms, Pie Charts

#### **Scatter**

##### **Creating Scatter Plots**

- With Pyplot, you can use the `scatter()` function to draw a scatter plot.
- The `scatter()` function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis.

**Example**

> Ploting a simple scatter plot.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

The observation in the example above is the result of 13 cars passing by.

> The X-axis shows how old the car is. <br>
> The Y-axis shows the speed of the car when it passes.

It seems that the newer the car, the faster it drives, but that could be a coincidence, after all we only registered 13 cars.

##### **Comparing Plots**

In the example above, there seems to be a relationship between speed and age, but what if we plot the observations from another day as well.

**Example**

> Drawing two plots on the same figure.


```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

By comparing the two plots, I think it is safe to say that they both gives us the same conclusion: *the newer the car, the faster it drives*.

##### **Size**

- We can change the size of the dots with the `s` argument.

Note: We have to make sure the array for sizes has the same length as the arrays for the x- and y-axis.

**Example**

> Setting your own size for the markers.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([10,30,700,20,500,1000,60,90,10,30,600,80,70])

plt.scatter(x, y, s=sizes)

plt.show()
```

##### **Alpha**

- We can adjust the transparency of the dots with the alpha argument.

**Example**

> Setting your own size for the markers.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([10,30,700,20,500,1000,60,90,10,30,600,80,70])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

#### **Bars**

##### **Creating Bars**

- With Pyplot, you can use the `bar()` function to draw bar graphs.

**Example**

> Drawing 4 bars.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

The `bar()` function takes arguments that describes the layout of the bars.

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]

plt.bar(x, y)
plt.show()
```

##### **Horizontal Bars**

- We can the bars to be displayed horizontally instead of vertically, use the `barh()` function.

**Example**

> Drawing 4 horizontal bars.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

##### **Bar Color**

- The `bar()` and `barh()` take the keyword argument color to set the color of the bars.

**Example**
> Drawing 4 green bars.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "green")
plt.show()
```

##### **Bar Width**

- The `bar()` takes the keyword argument **width** to set the width of the bars.

**Example**
> Drawing 4 very thin bars.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "green", width = 0.5)
plt.show()
```

##### **Bar Height**

- The `barh()` takes the keyword argument **height** to set the height of the bars.

**Example**
> Drawing 4 very thin bars.

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.3)
plt.show()
```

#### **Histograms**

A histogram is a graph showing frequency distributions. It is a graph showing the number of observations within each given interval.

##### **Creating Histogram**

- In Matplotlib, we use the `hist()` function to create histograms.

**Example**

> A Normal Data Distribution by NumPy.

```python
import numpy as np

x = np.random.normal(100, 10, 200)

print(x)
```

The `hist()` function will read the array and produce a histogram:

**Example**

> A simple histogram.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(100, 10, 200)

plt.hist(x)
plt.show()
```

#### **Pie Charts**

##### **Creating Pie Charts**

- With Pyplot, you can use the `pie()` function to draw pie charts.

**Example**

> Ploting a simple pie chart.

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

##### **Labels**

- The `label` parameter must be an array with one label for each wedge.

**Example**

> Ploting a simple pie chart with labels.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

##### **Explode**

- The `explode` parameter, if specified, and not None, must be an array with one value for each wedge.

**Example**

> Pulling the "Apples" wedge 0.2 from the center of the pie.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

##### **Colors**

- The `colors` parameter, if specified, must be an array with one value for each wedge.

**Example**

> Specify a new color for each wedge.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["aqua", "hotpink", "blue", "green"]

plt.pie(y, labels = mylabels, explode = myexplode, colors = mycolors)
plt.show()
```

##### **Legend**

- To add a list of explanation for each wedge, use the `legend()` function.

**Example**

> Adding a legend.

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["aqua", "hotpink", "blue", "green"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.legend(title = "Four Fruits")     # Add the title parameter
plt.show()
```
