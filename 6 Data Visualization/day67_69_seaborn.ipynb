#### **Seaborn üåäüî•**

Seaborn is built on top of Matplotlib and provides a higher-level interface for creating statistical graphics. While Matplotlib is a general-purpose plotting library, Seaborn is specifically designed for statistical data visualization.

#### **Usage of Seaborn**

- It is designed to work with Pandas dataframes, making it easy to visualize and explore data quickly and effectively.

- It offers a variety of powerful tools for visualizing data, including scatter plots, line plots, bar plots, heat maps, and many more.
- It also provides support for advanced statistical analysis, such as regression analysis, distribution plots, and categorical plots.

### **Seaborn Plot Types**

Seaborn provides a wide range of plot types that can be used for data visualization and exploratory data analysis.
  - Univariate ‚Äì x only (contains only one axis of information)
  - Bivariate ‚Äì x and y (contains two axis of information)
  - Trivariate ‚Äì x, y, z (contains three axis of information)

<br>

Here are some of the most commonly used plot types in Seaborn.

- Scatter Plot: A scatter plot is used to visualize the relationship between two variables.

- Line Plot: A line plot is used to visualize the trend of a variable over time.

- Histogram: A histogram is used to visualize the distribution of a variable.

- Box Plot: A box plot is used to visualize the distribution of a variable.

- Violin Plot: A violin plot is similar to a box plot, but provides a more detailed view of the distribution of the data.

- Heatmap: A heatmap is used to visualize the correlation between different variables.

- Pairplot: A pairplot is used to visualize the relationship between multiple variables.

#### **Scatter Plots**

- Scatter plots are used to visualize the relationship between two continuous variables.

- In Seaborn, scatter plots can be created using the `scatterplot()` function.

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
```

#### **Line Plots**

- Line plots are used to visualize trends in data over time or other continuous variables.

- In Seaborn, line plots can be created using the `lineplot()` function.

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal", data=fmri)
```

#### **Bar Plots**

- Bar plots are used to visualize the relationship between a categorical variable and a continuous variable.

- In Seaborn, bar plots can be created using the `barplot()` function.

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", data=titanic)
```

#### **Histograms**

- Histograms visualize the distribution of a continuous variable.

- In Seaborn, histograms can be created using the `histplot()` function.

```python
import seaborn as sns

iris = sns.load_dataset("iris")

sns.histplot(x="petal_length", data=iris)
```

#### **Density Plots**

- Density plots, also known as kernel density plots, are a type of data visualization that display the distribution of a continuous variable.

- In Seaborn, density plots can be created using the kdeplot() function.

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.kdeplot(data=tips, x="total_bill")
```

#### **Box Plots**

- Box plots are a type of visualization that shows the distribution of a dataset.

- In Seaborn, box plots can be created using the `boxplot()` function.

```python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)
```

#### **Violin Plots**

- A violin plot is a type of data visualization that combines aspects of both box plots and density plots into one.

- In Seaborn, violin plots can be created using the `violinplot()` function.

```python
import seaborn as sns

iris = sns.load_dataset("iris")

sns.violinplot(x="species", y="petal_length", data=iris)
```

#### **Heatmaps**

- A heatmap is a graphical representation of data that uses colors to depict the value of a variable in a two-dimensional space.

- In Seaborn, box plots can be created using the `heatmap()` function.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# # Loading the dataset
tips = sns.load_dataset("tips")

# Creating a heatmap of the correlation between variables
corr = tips.corr(numeric_only=True)
sns.heatmap(corr)

# Showing the plot
plt.show()
```

Another example of a heatmap using the `flights` dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
flights = sns.load_dataset('flights')

# Pivoting the data using keyword arguments
flights = flights.pivot(index='month', columns='year', values='passengers')

# Creating a heatmap
sns.heatmap(flights, cmap='Oranges', annot=True, fmt='d')

# Setting the title and axis labels
plt.title('Passengers per month')
plt.xlabel('Year')
plt.ylabel('Month')

# Showing the plot
plt.show()
```

#### **Pair Plots**

- Pair plots are a type of visualization in which multiple pairwise scatter plots are displayed in a matrix format.

- Each scatter plot shows the relationship between two variables, while the diagonal plots show the distribution of the individual variables.

- In Seaborn, box plots can be created using the `pairplot()` function.

```python
import seaborn as sns

# Loading the dataset
iris = sns.load_dataset("iris")

# Creating pair plot
sns.pairplot(data=iris)

# Showing the plot
plt.show()
```

We can customize the above plot by using `hue` and `diag_kind` parameter.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
iris = sns.load_dataset("iris")

# Creating pair plot with custom settings
sns.pairplot(data=iris, hue="species", diag_kind="kde", palette="husl")

# Setting the suptitle at the top middle
plt.suptitle("Iris Dataset Pair Plot", y=1.02, fontsize=15)

# Showing the plot
plt.show()
```

#### **Seaborn Facet Grids**

- FacetGrid is a powerful seaborn tool that us you to visualize the distribution of one variable as well as the relationship between two variables, across levels of additional categorical variables.

- FacetGrid creates a grid of subplots based on the unique values in the categorical variable specified.

```python
import seaborn as sns

# Loading the tips dataset
tips = sns.load_dataset('tips')

# Creating a FacetGrid for day vs total_bill
g = sns.FacetGrid(tips, col="day")

# Plotting histogram for total_bill in each day
g.map(sns.histplot, "total_bill")
```

#### **Customizing Seaborn Plots**

- Seaborn is a powerful data visualization library that provides numerous ways to customize the appearance of plots.

##### **Changing Color Palettes**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the tips dataset
tips = sns.load_dataset('tips')

# Creating a scatter plot with color palette
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, palette="Set2")

# Customizing plot
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

##### **Adjusting Figure Size**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
iris = sns.load_dataset("iris")

# Creating a violin plot with adjusted figure size
plt.figure(figsize=(8,6))
sns.violinplot(x="species", y="petal_length", data=iris)

# Customizing plot
plt.title("Petal Length Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

##### **Changing Color Palettes**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Loading the dataset
diamonds = sns.load_dataset("diamonds")

# Creating a scatter plot with annotations
sns.scatterplot(x="carat", y="price", data=diamonds)

# Adding annotations
plt.text(1, 18000, "Large, Expensive Diamonds", fontsize=12, color="red")
plt.text(2.5, 5000, "Small, Affordable Diamonds", fontsize=12, color="green")

# Customizing the plot
plt.title("Diamond Prices by Carat")
plt.xlabel("Carat (ct)")
plt.ylabel("Price ($)")
plt.show()
```
