### **Data Structure - Linked List üîóüîó**

Linked List is a linear data structure, in which elements are not stored at a contiguous location, rather they are linked using pointers. Linked List forms a series of connected nodes, where each node stores the data and the address of the next node.

linked_list.png

- Node Structure: A node in a linked list typically consists of two components:
- Data: It holds the actual value or data associated with the node.
- Next Pointer: It stores the memory address (reference) of the next node in the sequence.
- Head and Tail: The linked list is accessed through the head node, which points to the first node in the list. The last node in the list points to NULL or nullptr, indicating the end of the list. This node is known as the tail node.

Linked lists are necessary because they offer dynamic memory allocation, making it easy to insert or delete elements without shifting data. They efficiently utilize memory and allow the implementation of various data structures like stacks, queues, and graphs.

#### **Linked List - Types**

- Single-linked list: Each node points to the next node, traversing in a forward direction.
- Double-linked list: Each node has references to both the next and previous nodes, allowing forward and backward traversal.
- Circular linked list: The last node points back to the head, creating a circular structure.

#### **Linked List - Common Operations**

- Insertion: Adding a new node at the beginning, end, or any position within the list.
- Deletion: Removing a node from the list while adjusting pointers of neighboring nodes.
- Searching: Traversing the list from the head to find a specific value.

#### In this example, the `append(data)` method has a time complexity of **O(n)** - "linear time."
> As the linked list size (n) increases, the time taken for the `append` operation increases linearly. Each new element requires one additional step to find the last node and add the data.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Example usage:
if __name__ == "__main__":
    linked_list = LinkedList()
    linked_list.append(1)
    linked_list.append(2)
    linked_list.append(3)
    linked_list.display()
```
