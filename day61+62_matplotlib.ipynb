#### **Pyplot**

- Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

##### **Example**

> Draw a line in a diagram from position (0, 0) to position (5, 200)

```python
import numpy as np

xpoints = np.array([0, 5])      # The x-axis is the horizontal axis.
ypoints = np.array([0, 200])    # The x-axis is the vertical axis.

plt.plot(xpoints, ypoints)
plt.show()
```

#### **Plotting**

- The `plot()` function is used to draw points (markers) in a diagram. By default, the `plot() `function draws a line from point to point.

- Parameter 1 is an array containing the points on the **x-axis**. Parameter 2 is an array containing the points on the **y-axis**.

##### **Example**

> Draw a line in a diagram from position (0, 0) to position (5, 250).


```python
xpoints = np.array([0, 5])      # The x-axis is the horizontal axis.
ypoints = np.array([0, 250])    # The y-axis is the vertical axis.

plt.plot(xpoints, ypoints)
plt.show()
```

##### **Plotting Without Line**

- To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

##### **Example**

> Drawing two points in the diagram, one at position (1, 6) and one in position (5, 10).

```python
xpoints = np.array([1, 5])
ypoints = np.array([6, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

##### **Plotting Multiple Points**

- We can plot as many points as you like, just make sure you have the same number of points in both axis.

##### **Example**

> Drawing a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10).

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

##### **Default X-Points**

- If we do not specify the points on the x-axis, it will get the default values 0, 1, 2, 3 (etc.), depending on the length of the y-points.

##### **Example**

> Plotting without x-points. Here, the x-points in the example below are [0, 1, 2, 3, 4, 5].

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

#### **Markers**

- Marker is used to emphasize each point with a specification.

##### **Example**

> Marking each point with a **circle**.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = 'o')
plt.show()
```

##### **Example**

> Marking each point with a **star**.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, marker = '*')
plt.show()
```

##### **Formatting Strings**

- We can also use the shortcut string notation parameter to specify the marker.

- This parameter is also called `fmt`, and is written with this syntax: `marker|line|color`

##### **Marker Reference**

|Marker|Description|
|-----|-----|
|'o'|Circle|
|'*'|Star|
|'.'|Point|
|'s'|Square|
|'d'|Diamond (thin)|
|'X'|X (filled)|
|'P'|	Plus (filled)|

##### **Line Reference**

|Line Syntax|Description|
|-----|-----|
|'-'|	Solid line|
|':'|Dotted line|
|'--'|Dashed line|


##### **Color Reference**

|Color Syntax|Description|
|-----|-----|
|'r'|Red|
|'b'|Blue|
|'g'|Green|
|'c'|Cyan|
|'m'|Magenta|
|'k'|Black|

##### **Example**

> Marking each point with a circle with dashed line in cyan color.

```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, 'o--c')
plt.show()
```

##### **Marker Color**

- We can use the keyword argument `markerfacecolor` or the shorter `mfc` to set the color inside the edge of the markers:

##### **Example**

> Displaying the FACE color to red.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

##### **Example**

> Displaying the color of both the edge and the face to red.

Here, we use both the mec and mfc arguments to color the entire marker.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

##### **Example**

> Marking each point with a beautiful green color. We can use Hexadecimal color values.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')      # Using Hexadecimal color values
plt.show()
```

##### **Example**

> Marking each point with the color named "hotpink". We can also use [HTML Color Name](https://www.w3schools.com/colors/colors_names.asp)

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'HotPink')     # Using supported color names
plt.show()
```
