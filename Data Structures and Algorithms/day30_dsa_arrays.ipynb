### **DSA - Arrays (Python List) üìöüî¢**

Arrays known as lists in Python, are a fundamental data structure that allows us to store and organize a collection of elements of the same type. Arrays provide a way to efficiently manage and access multiple data values under a single variable name.

In real-life scenarios, arrays can be compared to containers or storage units that hold similar items together. Let's consider a few examples to understand the concept better:

1. **Grocery Shopping List:** Imagine you are going grocery shopping and you create a list of items you need to buy. Each item on the list represents an element in the array. You can add new items to the end of the list or remove items from it as you shop. The list allows you to keep track of all the items you need to purchase.

2. **Student Grades:** In a classroom, you may have a list of student grades for a particular subject. Each student's grade represents an element in the array. The array allows you to store and access the grades of all the students in a structured manner.

3. **Temperature Readings:** A weather station may collect temperature readings at regular intervals throughout the day. These readings can be stored in an array, where each element represents a temperature value at a specific time. The array allows for efficient retrieval and analysis of temperature data.

#### **Storage in Memory**

When an array is created in Python, a block of memory is allocated to store the elements. Each element is assigned a memory address based on its index in the array. This allows for efficient indexing and retrieval of elements. The memory allocation for arrays is flexible, allowing for the addition or removal of elements as needed.

#### **Big O Analysis ‚è≥üîç**

The time complexity of various operations on arrays can be analyzed using Big O notation:

- **Accessing an element by index:** O(1) - Just like accessing an item in your grocery shopping list by its position, retrieving an element from an array based on its index can be done in constant time.
- **Insertion/Deletion at the end:** O(1) - Adding or removing an item at the end of your grocery shopping list can be done in constant time since you simply update the list without shifting other items.
- **Insertion/Deletion at the beginning/middle:** O(n) - Adding or removing an item at the beginning or middle of your grocery shopping list requires shifting items to accommodate the change, resulting in a linear time complexity.
- **Searching for an item:** O(n) - Imagine you have an array of phone numbers and you want to find a specific number. Searching for it requires iterating through each number in the array, resulting in a linear time complexity.
- **Sorting an array:** O(n log n) - If you want to sort your grocery shopping list in alphabetical order, it would take an average time complexity of n log n using efficient sorting algorithms.

Understanding the properties and performance characteristics of arrays can help you solve real-life problems more effectively. By leveraging arrays, you can organize, manipulate, and analyze data in a structured and efficient manner.

##### **Exercise 1**
Expenses for every month are listed below, now let's create a list to store these monthly expenses and using that find out.

- January - 2200
- February - 2350
- March - 2600
- April - 2130
- May - 2190

```python
# Creating a Array (list) of monthly expenses
expenses = [2200, 2350, 2600, 2130, 2190]

# 1. In February, how many dollars you spent extra compare to January?
print("In February this much extra was spent compared to January:",expenses[1]-expenses[0]) # 150

# 2. Finding out the total expense in first quarter (first three months) of the year
print("Expense for first quarter:",expenses[0] + expenses[1]+ expenses[2]) # 7150

# 3. Finding out if you spent exactly 2000 dollars in any month
print("Did I spent 2000$ in any month?", 2000 in expenses) # False

# 4. June month just finished and my expense is 1980 dollar. Add this item to our monthly expense list
expenses.append(1980)
print("Expenses at the end of June:",expenses) # [2200, 2350, 2600, 2130, 2190, 1980]

# 5. Returning an item that you bought in a month of April and got a refund of $200. Make a correction to your monthly expense list based on this
expenses[3] = expenses[3] - 200
print("Expenses after 200$ return in April:", expenses) # [2200, 2350, 2600, 1930, 2190, 1980]
```
