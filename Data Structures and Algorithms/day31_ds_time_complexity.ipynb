### Time Complexity, Big O Analysis

### **Time Complexity ⏱️📈**

Time complexity refers to the measure of how the running time of an algorithm grows as the size of the input increases. It helps us understand how efficient an algorithm is in handling larger datasets.

In this example, the time complexity is **O(n)** - "linear time." As the array size (n) increases, the function's time taken grows linearly with it. So, the algorithm's performance scales directly with the input size.

```python
def sum_array(arr):
    total = 0
    for num in arr:         #
        total += num
    return total
```

#### **Big O notation ⬆️🅾️**

Big O notation, commonly denoted as **O()**, is a mathematical notation used to describe the time complexity or space complexity of an algorithm. It provides a way to analyze how the execution time or memory requirements of an algorithm grow as the size of the input data increases.

The Big O notation is usually used to describe the worst-case scenario.
Common Big O notations:

- **O(1)** - Constant Time: The algorithm's execution time or space requirement remains constant regardless of the input size. It means the algorithm has a constant efficiency, independent of the input.

- **O(log n)** - Logarithmic Time: The algorithm's efficiency grows logarithmically with the input size. As the input size increases, the time or space needed for the algorithm grows slowly.

- **O(n)** - Linear Time: The algorithm's efficiency grows linearly with the input size. The time or space needed increases in direct proportion to the input size.

- **O(n log n)** - Log-linear Time: The algorithm's efficiency grows in between linear and logarithmic time. Commonly seen in efficient sorting and searching algorithms like Merge Sort and Quick Sort.

- **O(n^2)** - Quadratic Time: The algorithm's efficiency grows quadratically with the input size. This is less efficient than linear time and is common in algorithms with nested loops.

- **O(n^k)** - Polynomial Time: The algorithm's efficiency grows polynomially with the input size, where k is a constant. Algorithms with nested loops leading to higher powers of n (e.g., O(n^3), O(n^4)) fall into this category.

- **O(2^n)** - Exponential Time: The algorithm's efficiency grows exponentially with the input size. These algorithms become extremely inefficient for large inputs.

- **O(n!)** - Factorial Time: The algorithm's efficiency grows with the factorial of the input size. This is the least efficient growth rate and is typically encountered in brute-force solutions to problems.

#### **Big O notation Analysis ⏳🔍**

```python
def main():
    n = int(input("Enter a number: "))
    count = 0

    for i in range(n):
      for j in range(n):
            count = count + 1

    print("n =", n, ", count =", count)

if __name__ == "__main__":
    main()
```

Here, the *time complexity* (Big O notation) of this code is **O(n^2)**.

> Explanation: The outer loop runs n times, and for each iteration of the outer loop, the inner loop also runs n times. Since both loops are directly dependent on the value of n, the total number of iterations will be  `n * n`, which is equivalent to `n^2`.

As n grows larger, the number of iterations increases quadratically. This means that the code's time complexity grows quickly with the size of n.


```python
def main():
    n = int(input("Enter a number: "))
    count = 0

    for i in range(n):
      for j in range(n):
        for k in range(n):
            count = count + 1

    print("n =", n, ", count =", count)

if __name__ == "__main__":
    main()
```

Here, the *time complexity* (Big O notation) of this code is **O(n^3)**.

> Because it uses 3 (three) nested loops, and the number of iterations is proportional to the cube of the input value `n`. As `n` increases, the execution time grows cubically.
